```python
data = censusdata.download("acs5", 2019, counties, list(tables.keys()), "eaaa132f157d69455a4e9d33c204db192eab407b", 'subject')
```

```python
sumCols = []

for column in data.columns[1:]:
    if tables[column] != 'no' and tables[column] != 'na':
        sumCols.append(column)
```

```python
essentialSums = pd.DataFrame(data[sumCols].sum(axis=1)).reset_index()
essentialSums = essentialSums.merge(data.reset_index()[['index','S2401_C01_001E']], on="index")
essentialSums.columns = ['county','essential_workers','total_workers']
essentialSums['pct_essential'] = essentialSums['essential_workers']/essentialSums['total_workers']
essentialSums['geoid'] = essentialSums['county'].apply(lambda x: parseCountyGEOID(x))
```
